```python
def input_data():
    data = list(map(int, input("Enter at least 10 numbers (space-separated): ").split()))
    if len(data) < 10:
        print("Please enter at least 10 numbers.")
        return input_data()
    return data
```

```python
def factorial(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f
```

```python
def transform_data(data):
    print("\n1. Square\n2. Cube\n3. Factorial")
    choice = int(input("Choose transform: "))
    
    if choice == 1:
        return [x**2 for x in data]
    elif choice == 2:
        return [x**3 for x in data]
    elif choice == 3:
        return [factorial(x) for x in data]
    else:
        print("Invalid choice")
        return data
```

```python
def filter_data(data):
    limit = int(input("Show values greater than: "))
    return [x for x in data if x > limit]
```

```python
def sort_data(data):
    print("1. Ascending\n2. Descending")
    choice = int(input("Choose: "))
    
    if choice == 1:
        return sorted(data)
    else:
        return sorted(data, reverse=True)
```

```python
def statistics(data):
    print("\n--- Statistics ---")
    print("Max:", max(data))
    print("Min:", min(data))
    print("Sum:", sum(data))
```

```python
data = input_data()

while True:
    print("\nMENU")
    print("1. Transform Data")
    print("2. Filter Data")
    print("3. Sort Data")
    print("4. Statistics")
    print("5. Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        data = transform_data(data)
        print("Updated Data:", data)

    elif ch == 2:
        print("Filtered Data:", filter_data(data))

    elif ch == 3:
        data = sort_data(data)
        print("Sorted Data:", data)

    elif ch == 4:
        statistics(data)

    elif ch == 5:
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```
